```julia
function fastcholesky_tridiag!(dv, ev)
    error(3)
    
    a = dv
    b = ev
    
    # @inbounds dv[1] = a[1]
    @inbounds for i in 2:length(dv)
        # dv[i] = a[i] - (b[i - 1])^2 / (dv[i - 1])
        dv[i] = max(a[i] - (b[i - 1])^2 / (dv[i - 1]), 0.01)
    end
    
    dv = map!(sqrt, dv, dv)
    ev = map!(/, ev, ev, dv)

    return Cholesky(Bidiagonal(dv, ev, :L), 'L', 0)
end

function fastcholesky_tridiag(dv, ev)
    error(4)
    a = ReverseDiff.value(dv)
    b = ReverseDiff.value(ev)
    
    ndv = zeros(length(dv))
    nev = zeros(length(ev))
    
    @inbounds ndv[1] = a[1]
    @inbounds for i in 2:length(dv)
        ndv[i] = a[i] - (b[i - 1])^2 / (ndv[i - 1])
        # ndv[i] = max(a[i] - (b[i - 1])^2 / (ndv[i - 1]), 0.01)
    end
    
    ndv = sqrt.(ndv)
    nev = nev ./ view(ndv, firstindex(nev):lastindex(nev))

    return Cholesky(Bidiagonal(ndv, nev, :L), 'L', 0)
end
```

```julia
structured(model::DynamicPPL.Model) = structured(Random.default_rng(), model)

function structured(rng::Random.AbstractRNG, model::DynamicPPL.Model)
    # Setup.
    varinfo = DynamicPPL.VarInfo(model)
    num_params = length(varinfo[DynamicPPL.SampleFromPrior()])

    # initial params
    l = 3num_params - 1
    #  = randn(l)
    # , s, offs = ftransform()
    
     = zeros(num_params)
    s = 1softplus.(ones(num_params))
    offs = -0.1 * ones(num_params - 1)
    
    # C = fastcholesky_tridiag(s, offs)
    L = Bidiagonal(s, offs, :L)
    C = Cholesky(L, 'L', 0)

    # Construct the base family.
    # d = DistributionsAD.TuringDiagMvNormal(, )
    # d = DistributionsAD.TuringDiagMvNormal(, )
    d = Turing.DistributionsAD.TuringDenseMvNormal(, C)

    # Construct the bijector constrained  unconstrained.
    b = Bijectors.bijector(model; varinfo=varinfo)

    # We want to transform from unconstrained space to constrained,
    # hence we need the inverse of `b`.
    return Bijectors.transformed(d, Bijectors.inverse(b))
end

# Overloading stuff from `AdvancedVI` to specialize for Turing
function AdvancedVI.update(d::Turing.DistributionsAD.TuringDenseMvNormal, , , off)
    # M = fastcholesky_tridiag(, off)
    L = Bidiagonal(, off, :L)
    M = Cholesky(L, 'L', 0)
    return Turing.DistributionsAD.TuringDenseMvNormal(, M)
end
    
function AdvancedVI.update(
    td::Bijectors.TransformedDistribution{<:Turing.DistributionsAD.TuringDenseMvNormal},
    ::AbstractArray,
)
    # , s, offs = ftransform()
    l = length(td)
     = [1:l]
    s, offs = softplus.([(l + 1):2l]), [(2l + 1):end]
    return AdvancedVI.update(td, , s, offs)
end

function AdvancedVI.vi(
    model::DynamicPPL.Model,
    alg::AdvancedVI.ADVI,
    q::Bijectors.TransformedDistribution{<:Turing.DistributionsAD.TuringDenseMvNormal};
    optimizer = AdvancedVI.TruncatedADAGrad(),
)
    # Initial parameters for structured approx            
    #  = btransform(q)
    num_params = length(q)
     = vcat(1.0, collect(Iterators.flatten(states))) # randn(num_params)
    #  = 3randn(num_params)
    prior = ones(num_params - 1)
    for i in 1:T
        prior[2 + (i - 1) * 4] = observations[i]
    end        
     = vcat(1.0, prior)
    s = 1softplus.(ones(num_params))
    offs = -1 * ones(num_params - 1)
     = vcat(, s, offs)

    # Optimize
    # AdvancedVI.optimize!(elbo, alg, q, Turing.Variational.make_logjoint(model), ; optimizer = optimizer)
    # Copypasted from Turing.jl but added callbacks and checks
    vo = ELBO()
        
    samples_per_step = alg.samples_per_step
    max_iters = alg.max_iters

    diff_result = DiffResults.GradientResult()

    i = 0
    progress = ProgressMeter.Progress(max_iters, 1, "Optimizing...", 0)
        
    log = Turing.Variational.make_logjoint(model)

    rng = StableRNG(123)
    es = []

    # add criterion? A running mean maybe?
    time_elapsed = @elapsed while (i < max_iters) # & converged
        AdvancedVI.grad!(vo, alg, q, log, , diff_result, samples_per_step)

        # apply update rule
         = DiffResults.gradient(diff_result)
         = AdvancedVI.apply!(optimizer, , )
        @.  =  - 
        
        ProgressMeter.next!(progress)
            
        q = AdvancedVI.update(q, )
        
        push!(es, vo(rng, alg, q, log, samples_per_step))

        i += 1
    end

    # Return updated `Distribution`
    return es, AdvancedVI.update(q, )
end
```
